# Tiny PCA-MLP run

Launch a small PCA-MLP run and view its RMSEs


```python
from pathlib import Path
import pandas as pd
from thousandworlds.run_model import run

data_dir = Path("../dataset") if Path("../dataset").exists() else Path("dataset")
```

```python
result = run(
    "pca_mlp",
    "single-complete",
    data_dir=data_dir,
    ppca_iters=10,
    num_steps=20,
    latent_dim=8,
    hidden_width=16,
    dtype="float32",
)
```

```python
pd.Series(result["metrics"]["rmse"]["per_variable"], name="rmse")
```
